Create a pandas dataframe(DataFrame as 'df') with numpy random values (4 features and 4 observations.


```python
import pandas as pd
import numpy as np
Features=np.random.rand(4)
Observations=np.random.rand(4)
df=pd.DataFrame([Features,Observations])
df
```
